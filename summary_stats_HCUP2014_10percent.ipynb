```python
dataset['AGE'].plot(kind='hist',bins=91,density=True,edgecolor='black')
plt.xlabel('age')
plt.title('Patient age distribution')
plt.show()
nan_counts=dataset['AGE'].isnull().sum()
print('There are '+str(nan_counts)+' missing values, which account for '+str(nan_counts/len(dataset)*100)+'%')
```
